# data clean & pre-processing 
dealing with missing data

```python
titanic.isna().sum()  #returns the number of missing values in each column.
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())  #to fill missing value
```

```python
titanic.isna().sum()
```

```python
titanic
```

```python
titanic.drop("Cabin" , axis=1 , inplace=True ) #to remove this column from table
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic['Sex'].value_counts()   # to determine number of male and female
```

```python
titanic['Embarked'].value_counts()   # to determine number of embarked
```

```python
sex = pd.get_dummies(titanic['Sex'] , drop_first=True)
sex
```

```python
emb = pd.get_dummies(titanic['Embarked'] )
emb
```

```python
titanic=pd.concat([titanic,sex,emb],axis=1)
titanic
```

```python
titanic.drop(['PassengerId' ,'Name' , 'Sex' , 'Ticket' ,'Embarked'] , axis = 1 , inplace=True)
```

```python
titanic
```

```python
titanic.rename(columns={'male': 'Gender'},inplace=True)  #to change name of sthings
titanic
```
